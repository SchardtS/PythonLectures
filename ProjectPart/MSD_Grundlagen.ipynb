```python
import numpy as np
import matplotlib.pyplot as plt
```

## Funktionen definieren

```python
def random_walk(T = 100,dt = 0.01):
    nTimeSteps = int(T/dt)
    x = np.zeros((nTimeSteps,2))
    for i in range(1,  nTimeSteps):
         x[i] = x[i-1] + np.sqrt(dt)*np.random.randn(1,2)   
    return x
```

```python
def squaredNorm(x):
    N = len(x)
    sNorm = np.zeros(N)
    for t in range(N):
        sNorm[t] = (np.linalg.norm(x[t] - x[0]))**2
    return sNorm
```

## Ein Agent

```python
trajectory=random_walk(T = 5,dt = 0.01)
```

```python
plt.scatter(trajectory[:,0],trajectory[:,1])
plt.xlabel("x-coordinate")
plt.ylabel("y-coordinate")
plt.title("trajectory of one agent")
plt.show()
```

## Mehrere Agenten

```python
for i in range(5):
    trajectory=random_walk(T = 5,dt = 0.01)
    
    plt.scatter(trajectory[:,0],trajectory[:,1])
    plt.xlabel("x-coordinate")
    plt.ylabel("y-coordinate")
    plt.title("trajectories of several agents")

plt.show()
```

## Mittlere quadratische Verschiebung (Mean squared displacement - MSD)

Die mittlere quadratische Verschiebung (Mean squared displacement - MSD) zum Zeitpunkt $t$ is definiert als
$$MSD(t)=\frac{1}{n}\sum_{i=1}^n|x_i(t)-x_i(0)|^2,$$ 
wobei $n$ die Anzahl der Agenten ist, der Vektor $x_i(t)$ die Position von Agent $i$ zum Zeitpunkt $t$ und der Vektor $x_i(0)$ die Position von Agent $i$ zum Zeitpunkt $0$.

```python
T = 5
dt = 0.01
sNorm_ensemble = np.zeros(int(T/dt))
numAgents = 200
for i in range(numAgents):
    trajectory=random_walk(T = T,dt = dt)
    sNorm_ensemble += squaredNorm(trajectory)
MSD = sNorm_ensemble/numAgents
plt.plot([i*dt for i in range(int(T/dt))],MSD)
plt.xlabel("time")
plt.ylabel("Ensemble MSD")
plt.show()
```

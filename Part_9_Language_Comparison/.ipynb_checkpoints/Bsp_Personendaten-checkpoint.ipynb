```r
library(ggplot2)

data <- read.csv('PersonenDaten.csv', na='NA')

ggplot( data, aes(x = Schuhgroesse))+
    geom_histogram( binwidth=1, color='black', fill='grey' )+
    geom_vline( aes( xintercept = mean(Schuhgroesse), color='Mittelwert' ), size=1 )+
    geom_vline( aes( xintercept = median(Schuhgroesse), color='Median' ), size=1 )+
    scale_color_manual( "LagemaÃŸe",
                         breaks = c("Mittelwert", "Median"),
                         values = c('magenta', 'blue') )+
    ggtitle('Verteilung Schuhgroessen')+
    labs(y='Anzahl')+
    theme(aspect.ratio=0.5)
  
```
